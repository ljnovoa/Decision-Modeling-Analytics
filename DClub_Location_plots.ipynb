The plots in this notebook are based on the "DClub" example seen in class (Source: Optimization In Operations Research by Rardin) <br>
The Matplotlib code for plotting surfaces and contours below, is based on very nice content available at: <br> https://www.youtube.com/watch?v=gqoLLGgbeAE&ab_channel=KimberlyFessel and <br> https://matplotlib.org/stable/gallery/images_contours_and_fields/contour_demo.html

```python
#See https://numpy.org/doc/stable/
import numpy as np
#See https://matplotlib.org/stable/tutorials/index  and  https://matplotlib.org/stable/tutorials/introductory/pyplot.html
import matplotlib.pyplot as plt 
```

```python
#Definition of Objective Function - the one we want to plot
def dclub_objective_function(x,y):
    return ((60/(1+(x+1)**2+(y-3)**2))+(20/(1+(x-1)**2+(y-3)**2))+(30/(1+(x)**2+(y+4)**2)))
```

```python
#Grid layout
#Here we're using N points along the x and y axes
N = 5000
#Use linspace (numpy - https://numpy.org/doc/stable/reference/generated/numpy.linspace.html) to generate N intervals between the first and second arguments 
x_points = np.linspace(-6, 6, N)
y_points = np.linspace(-7, 7, N)
plt.rcParams['figure.figsize'] = (10,10)
```

```python
#Use numpy's "meshgrid" function to define the grid using the points generated above (establishes the 2D domain)
X, Y = np.meshgrid(x_points, y_points)
#Evaluate every point in the grid in the DClub function
Z = dclub_objective_function(X, Y)
```

```python
#Create axes with the axes function (projection keyword specifies that the plot is 3D) and named the object "dclubax"
dclubax = plt.axes(projection = '3d')
dclubax.set_title('DClub Objective Function')
#Create labels for the axes using the "set" method (for ease of visualization)
dclubax.set_xlabel('x')
dclubax.set_ylabel('y')
dclubax.set_zlabel('Business Attracted')
#Generate the 3D plor using the "plot_surface" method
dclubax.plot_surface(X, Y, Z)
plt.show()
```

```python
#Creates figure and axes - see: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
LCfig, LCax = plt.subplots()
LCax.set_title('DClub Level Curves')
levelcurves = LCax.contour(X, Y, Z, 10) #last parameter corresponds to the number of levels in the contour plot
put_marks_at = [(-5, 0), (-3, 4), (-1, 4.5), (0, 3.5)]
#Labels on contour plot - %First#.Second#f format, First: min number of digits in the displayed string and Second: digits diplayed after the decimal point
LCax.clabel(levelcurves, inline=True, fontsize=9, manual=put_marks_at, fmt='%1.1f')
plt.show()
```
