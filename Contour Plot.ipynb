# Lemonade Stand - Contour Plot

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = [8, 5]
```

```python
p = 0.5
FC = 50
```

```python
def P(Q,c):
    return Q*(p-c) - FC
```

```python
fig = plt.figure()
x, y =  np.array([i for i in range(0,550,50)]),  np.array([i/10.0 for i in range(0,7,1)])
Q, c = np.meshgrid(x, y)
Z = P(Q, c)

ax = plt.axes(projection='3d')
ax.plot_surface(Q, c, Z, rstride=1, cstride=1, alpha=0.5)
#cset = ax.contourf(Q, c, Z, zdir='z', offset=-100, cmap=cm.Oranges)      #Constant P
#cset = ax.contourf(Q, c, Z, zdir='x', offset=0, cmap=cm.Oranges)          #Constant Q
#cset = ax.contourf(Q, c, Z, zdir='y', offset=0, cmap=cm.Oranges)         #Constant c
ax.set_xlabel('Q')
ax.set_xlim(0, 500)
ax.set_ylabel('c')
ax.set_ylim(0,0.6)
ax.set_zlabel('P(Q,c)')
ax.set_zlim(-100, 250)

ax.view_init(30, 30)
```
