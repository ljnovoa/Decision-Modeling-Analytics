# BSAN391 -- Optimal Pricing-Curve Fitting

```python
# Appropriate imports
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
# Define the desired functional form
def FunctionalForm(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
# Data
x = np.array([80,90,100,110,120,130,140,150,160,170,180,190,200,210,220,230,240])
y = np.array([-432000,-308000,-200000,-108000,-32000,28000,72000,100000,112000,108000,88000,52000,0,-68000,-152000,-252000,-368000])
```

```python
# Fit --> popt: array of optimal values for the parameters so that the sum of the squared residuals (f(x, *popt) - y)^2 is minimized.
popt = curve_fit(FunctionalForm, x, y)[0]
```

```python
plt.scatter(x, y, label='data')
plt.plot(x, FunctionalForm(x, *popt), 'g--',
         label='fit: a=%5.1f, b=%5.1f, c=%5.1f' % tuple(popt))
plt.xlabel('p')
plt.ylabel('Profit(p)')
plt.legend()
plt.show()
```
