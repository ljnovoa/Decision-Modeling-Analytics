# BSAN391 - Optimal Pricing

```python
# Libraries imports
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
# Parameters
x0 = 200       # For initializing the nonlinear optimization method, a staring point in the search
```

The profit as a function of the price $p$ can be written as: 

$$\text{profit}({\color{red}p})=-80{\color{red}p}^{2}+26000{\color{red}p}-2000000$$

We could define this function in Python as follows:

```python
# Profit Quadratic Function
def profit(p):
    return -1*(-80*p**2+26000*p-2000000)
```

Note that the function above returns the negative of the profit, as the method below minimizes by default:

```python
# Optimization Method                            
optimization = minimize(profit, x0, method='Nelder-Mead', options={'xtol': 1e-8, 'disp': False})
```

```python
optimization
```

The Nealder-Mead method searches without the use of derivatives, instead it keeps a list of solutions ranked by solution quality

```python
# Results Display - Two possible ways of displaying the results
if optimization.success:
    print(optimization.message)
    print ("Optimal Price:", "$"+str(optimization.x[0]), "-- Maximum Profit:", "$"+str(-1*optimization.fun))
    print (f"Optimal Price: ${optimization.x[0]} -- Maximum Profit: ${-1*optimization.fun}") #See https://www.freecodecamp.org/news/python-f-strings-tutorial-how-to-use-f-strings-for-string-formatting/
else:
    print(optimization.message)
```

```python
# Plot
r = [i for i in range(80,241,1)] #List with integers from 80 to 240
plt.plot(r, [-profit(j) for j in r])
plt.xlabel('Price')
plt.ylabel('Profit')
plt.show()
```
